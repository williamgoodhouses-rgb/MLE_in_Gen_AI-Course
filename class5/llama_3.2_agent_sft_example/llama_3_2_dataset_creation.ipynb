```python
# Install the necessary libraries
! pip install datasets pandas

# Import the libraries
from datasets import load_dataset
import pandas as pd
```

```python

# Load the 'openai/MMMLU' dataset
dataset = load_dataset("openai/MMMLU")
```

## Exploring the MMLU Dataset

```python
df = dataset['test'].to_pandas()
```

```python
df
```

```python
df_chinese = df.tail(14000)
```

```python
df_chinese['Subject'].value_counts()
```

## Creating the Chinese Law dataset

```python
df_law = df_chinese[(df_chinese['Subject'] == 'professional_law') | (df_chinese['Subject'] == 'jurisprudence')]
df_law.shape
```

```python
# prompt: for the df_law dataframe I want to keep the column 'Question' and we create a new column with the column value that is mentionned under 'Answer' column example for the first line the the Answer column value is D then we keep the value of the D column in this new column we name 'Reply'

df_law['Reply'] = df_law.apply(lambda row: row[row['Answer']], axis=1)
```

```python
pd.set_option('display.max_colwidth', None)  # Set to None to display the full content of the columns
```

```python
df_law=df_law[['Question','Reply']]
```

```python
df_law.sample(5)
```

```python
df_law.to_csv('law.csv', index=False)
```

## Creating the Chinese Medicine dataset

```python
df_medicine = df_chinese[(df_chinese['Subject'] == 'professional_medicine') | (df_chinese['Subject'] == 'clinical_knowledge') | (df_chinese['Subject'] == 'college_medicine') | (df_chinese['Subject'] == 'medical_genetics')]
df_medicine.shape
```

```python
# Create a copy to avoid SettingWithCopyWarning
df_medicine = df_medicine.copy()

# For the df_medicine dataframe, keep the 'Question' column and create a new column 'Reply'
# The 'Answer' column contains the letter (A, B, C, or D) indicating which option is correct
# We extract the corresponding answer text from columns A, B, C, or D
df_medicine['Reply'] = df_medicine.apply(lambda row: row[row['Answer']], axis=1)
```

```python
df_medicine=df_medicine[['Question','Reply']]
```

```python
df_medicine.sample(5)
```

```python
df_medicine.to_csv('medicine.csv', index=False)
```
